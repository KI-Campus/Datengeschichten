# Aufbereitung
Jetzt ist es sinnvoll, dass wir uns die Variablen anschauen und ggfs. Änderungen vornhemen. Wie lässt sich das mit Hilfe von R lösen?

_Lösung:_
Wir betrachten zunächst die Klassifierzung der einzlenen Variablen und klassifizieren diese anschließend Schritt für Schritt in die für unsere Visualisierung benötigten Form um. Anschließend erstellen wir einen neuen Datensatz, der unsere so aufbereiteten Variablen enthält. 
Die Variable Geburtstjahr können wir auch in ihrer Klassifizierung verändern, allerdings benötigen wir sie für die weitere Visualisierung nicht, deswegen verzichten wir darauf sie mit in den neuen Datensatz zu schreiben.


# Aufbereitung
Wir müssen jetzt überlegen, ob die Daten in der Form vorliegen mit der wir unsere Bevölkerungspyramide erstellen können oder ob weitere Aufbereitungsschritte notwendig sind. Welche Aufbereitung ist an dieser Stelle ggfs. noch notwendig?

```r
#Aggregation auflösen
MI_Freq_Alter <- rep(MI_Aufbereitet$Alter_num,MI_Aufbereitet$gesamt_num) 
summary(MI_Freq_Alter)
table(MI_Freq_Alter)

MI_Freq_maennlich <- rep(MI_Aufbereitet$maennlich_num, MI_Aufbereitet$gesamt_num) 
summary(MI_Freq_maennlich)
table(MI_Freq_maennlich)


MI_Freq_weiblich <- rep(MI_Aufbereitet$weiblich_num, MI_Aufbereitet$gesamt_num) 
summary(MI_Freq_weiblich)
table(MI_Freq_weiblich)

data_2020 <- data.frame(MI_Freq_Alter, MI_Freq_maennlich, MI_Freq_weiblich)
```

_Lösung:_ 
Wie bei dem Datensatz mit dem wir die Bevölkerungspyramiden für die Stadt Bielefeld erstellt haben ist es notwendig die Aggregation der Daten aufzulösen. Wir multiplizieren hierfür die Häufigkeit der Mindener:innen, die in unserer Variable "gesamt_num" enthalten ist, jeweils mit unsereren beiden aufbereiteteten Geschlechter-Variablen sowie der Alters-Variablen . Das sind somit 3 weitere Vektoren die wir erstellen müssen. Der rep() Befehl gibt hierbei anhand der Häufigkeit der Mindener:innen das jeweilige Element(Alter/Geschlecht: männlich/weiblich) in entsprechender Anzahl zurück. Die so entstandenen Objekte werden zum Schluss in einen neuen Datensatz geschrieben.

# Bevölkerungspyramide
Jetzt können wir die Bevölkerungspyramide der Stadt Minden für das Jahr 2020 erstellen. Da wir die Daten für lediglich ein Jahr (2020) vorliegen haben und die beiden Geschlechtergruppen als einzelne Variablen bzw. Vektoren vorliegen können wir darauf verzichten Subsets anzulegen und direkt mit der Visualisierung starten. Hier können wir auf das Baukastenprinzip zurückgreifen mit dem es möglich ist Visualisierungen anzupassen und auszugestalten. 
Wie sieht unsere Bevölkerungspyramide aus?

```r
#Pyramide Minden
par(bg = "gray80") 
barplot(table(MI_Freq_weiblich, MI_Freq_Alter), axes = FALSE,
        horiz = TRUE, axis.lty = -1, border = NA, 
        xlim =c(-1600,1600), col = "firebrick3")
        grid (16,0, lty = 6, col = "cornsilk2")
barplot(-table(MI_Freq_maennlich, MI_Freq_Alter), axes =FALSE, horiz = TRUE, axis.lty = -1, border = NA,
        xlim =c(-1600,1600), col = "deepskyblue3", add=T)
        grid (16,0, lty = 6, col = "cornsilk2")
abline(v=0,lwd=0,col=par("bg"), font = 4)
for (i in seq(10,105,by=10)) text(0,i+i*0.2,i,cex=1.0)
mtext(abs(seq(-1600,1600,by=200)),at=seq(-1600,1600,by=200),1,line=0,cex=0.9, las=2)

#Beschriftung Links - Maenner
mtext("Männer",3,line=0,adj=0,cex=1.3)

#Beschriftung Rechts - Frauen
mtext("Frauen",3,line=0,adj=1,cex=1.3)

#Ueberschrift
mtext("Bevölkerungspyramide Minden 2020",
      3,line=-1.5,adj=0,cex=1.5,font=4,outer=T)

#Untertitel
mtext("(Datenquelle: Stadt Minden)",
      3,line=-2.5,adj=0,cex=1,font=3,outer=T)

#X-Achse: Beschriftung
mtext("Mindener:innen",
      1,line=-3.25,adj=0.55,cex=0.95,font=2,outer=T)

#y-Achse Beschriftung
mtext("Alter in Jahren*",
      3,line=-4,adj=0.55 ,cex=0.95,font=2,las=1,outer=T)
```

_Lösung:_ 
Es sind Anpassung in Bezug auf die vorliegenden Daten notwenig. Wir müssen die Skalierungen der Achsen anpassen und die Beschrifttungen ausrichten. Auch die Sequenz der Hilfslinien, sofern wir diese einzeichnen lassen wollen müssen wir entsprechend anpassen. Ansonsten ist die Erstellung der Bevölkerungspyramide für die Stadt Minden relativ ähnlich zur Erstellung der Pyramiden für die Stadt Bielefeld.
